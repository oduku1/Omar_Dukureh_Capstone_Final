Data Loading and Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
#load the data
Nba2018 = pd.read_csv('../data/raw/NBA2018.csv')
Nba2019 = pd.read_csv('../data/raw/NBA2019.csv')
Nba2020 = pd.read_csv('../data/raw/NBA2020.csv')
Nba2021 = pd.read_csv('../data/raw/NBA2021.csv')
Nba2022 = pd.read_csv('../data/raw/NBA2022.csv')
Nba2023 = pd.read_csv('../data/raw/NBA2023.csv')
Nba2024 = pd.read_csv('../data/raw/NBA2024.csv')

Nba2022.drop(columns=['Unnamed: 28','Unnamed: 27'], inplace=True)
Nba2023.drop(columns=['Unnamed: 28','Unnamed: 27'], inplace=True)
```

```python
#Load Advanced Stats Data
Nba2018ADV = pd.read_csv("../data/raw/2018ADV.csv")
Nba2019ADV = pd.read_csv("../data/raw/2019ADV.csv")
Nba2020ADV = pd.read_csv("../data/raw/2020ADV.csv")
Nba2021ADV = pd.read_csv("../data/raw/2021ADV.csv")
Nba2022ADV = pd.read_csv("../data/raw/2022ADV.csv")
Nba2023ADV = pd.read_csv("../data/raw/2023ADV.csv")
Nba2024ADV = pd.read_csv("../data/raw/2024ADV.csv")
```

```python
#Add year column to the dataframes

Nba2018["Year"] = 2018
Nba2019["Year"] = 2019
Nba2020["Year"] = 2020
Nba2021['Year'] = 2021
Nba2022['Year'] = 2022
Nba2023['Year'] = 2023
Nba2024['Year'] = 2024


Nba2018ADV['Year'] = 2018
Nba2019ADV['Year'] = 2019
Nba2020ADV['Year'] = 2020
Nba2021ADV['Year'] = 2021
Nba2022ADV['Year'] = 2022
Nba2023ADV['Year'] = 2023
Nba2024ADV['Year'] = 2024
```

```python
#dropNA values
Nba2018 = Nba2018.dropna()
Nba2019 = Nba2019.dropna()
Nba2020 = Nba2020.dropna()
Nba2021 = Nba2021.dropna()
Nba2022 = Nba2022.dropna()
Nba2023 = Nba2023.dropna()
Nba2024 = Nba2024.dropna()  

Nba2018ADV = Nba2018ADV.dropna()
Nba2019ADV = Nba2019ADV.dropna()
Nba2020ADV = Nba2020ADV.dropna()
Nba2021ADV = Nba2021ADV.dropna()
Nba2022ADV = Nba2022ADV.dropna()
Nba2023ADV = Nba2023ADV.dropna()
Nba2024ADV = Nba2024ADV.dropna()  
```

```python

# Sort by Player name and Games played (G) descending
Nba2018_sorted = Nba2018.sort_values(by=["Player", "G"], ascending=[True, False])

# Drop duplicates by Player, keeping the row with the most games
Nba2018_cleaned = Nba2018_sorted.drop_duplicates(subset="Player", keep="first")

Nba18 = Nba2018_cleaned.set_index("Player")

Nba2018ADV_sorted = Nba2018ADV.sort_values(by=["Player", "G"], ascending=[True, False])
Nba2018ADV_cleaned = Nba2018ADV_sorted.drop_duplicates(subset="Player", keep="first")
Nba18ADV = Nba2018ADV_cleaned.set_index("Player")


# Sort by Player name and Games played (G) descending
Nba2019_sorted = Nba2019.sort_values(by=["Player", "G"], ascending=[True, False])
# Drop duplicates by Player, keeping the row with the most games
Nba2019_cleaned = Nba2019_sorted.drop_duplicates(subset="Player", keep="first")
Nba19 = Nba2019_cleaned.set_index("Player")


Nba2019ADV_sorted = Nba2019ADV.sort_values(by=["Player", "G"], ascending=[True, False])
Nba2019ADV_cleaned = Nba2019ADV_sorted.drop_duplicates(subset="Player", keep="first")
Nba19ADV = Nba2019ADV_cleaned.set_index("Player")

# Sort by Player name and Games played (G) descending
Nba2020_sorted = Nba2020.sort_values(by=["Player", "G"], ascending=[True, False])
# Drop duplicates by Player, keeping the row with the most games
Nba2020_cleaned = Nba2020_sorted.drop_duplicates(subset="Player", keep="first")
Nba20 = Nba2020_cleaned.set_index("Player")

Nba2020ADV_sorted = Nba2020ADV.sort_values(by=["Player", "G"], ascending=[True, False])
Nba2020ADV_cleaned = Nba2020ADV_sorted.drop_duplicates(subset="Player", keep="first")
Nba20ADV = Nba2020ADV_cleaned.set_index("Player")


# Sort by Player name and Games played (G) descending
Nba2021_sorted = Nba2021.sort_values(by=["Player", "G"], ascending=[True, False])
# Drop duplicates by Player, keeping the row with the most games
Nba2021_cleaned = Nba2021_sorted.drop_duplicates(subset="Player", keep="first")
Nba21 = Nba2021_cleaned.set_index("Player")


Nba2021ADV_sorted = Nba2021ADV.sort_values(by=["Player", "G"], ascending=[True, False])
Nba2021ADV_cleaned = Nba2021ADV_sorted.drop_duplicates(subset="Player", keep="first")
Nba21ADV = Nba2021ADV_cleaned.set_index("Player")


Nba2022_sorted = Nba2022.sort_values(by=["Player", "G"], ascending=[True, False])
# Drop duplicates by Player, keeping the row with the most games
Nba2022_cleaned = Nba2022_sorted.drop_duplicates(subset="Player", keep="first")
Nba22 = Nba2022_cleaned.set_index("Player")



Nba2022ADV_sorted = Nba2022ADV.sort_values(by=["Player", "G"], ascending=[True, False])
Nba2022ADV_cleaned = Nba2022ADV_sorted.drop_duplicates(subset="Player", keep="first")
Nba22ADV = Nba2022ADV_cleaned.set_index("Player")



Nba2023_sorted = Nba2023.sort_values(by=["Player", "G"], ascending=[True, False])
# Drop duplicates by Player, keeping the row with the most games
Nba2023_cleaned = Nba2023_sorted.drop_duplicates(subset="Player", keep="first")
Nba23 = Nba2023_cleaned.set_index("Player")


Nba2023ADV_sorted = Nba2023ADV.sort_values(by=["Player", "G"], ascending=[True, False])
Nba2023ADV_cleaned = Nba2023ADV_sorted.drop_duplicates(subset="Player", keep="first")
Nba23ADV = Nba2023ADV_cleaned.set_index("Player")



Nba2024_sorted = Nba2024.sort_values(by=["Player", "G"], ascending=[True, False])
# Drop duplicates by Player, keeping the row with the most games
Nba2024_cleaned = Nba2024_sorted.drop_duplicates(subset="Player", keep="first")
Nba24 = Nba2024_cleaned.set_index("Player")


Nba2024ADV_sorted = Nba2024ADV.sort_values(by=["Player", "G"], ascending=[True, False])
Nba2024ADV_cleaned = Nba2024ADV_sorted.drop_duplicates(subset="Player", keep="first")
Nba24ADV = Nba2024ADV_cleaned.set_index("Player")
```

```python
Nba_combined = pd.concat([Nba18,Nba19,Nba20,Nba21,Nba22,Nba23, Nba24])

Nba_combinedADV = pd.concat([Nba18ADV,Nba19ADV,Nba20ADV,Nba21ADV,Nba22ADV,Nba23ADV, Nba24ADV])
#sort by player
Nba_combined = Nba_combined.sort_values(by=["Player"])


#reset index for joining
Nba_combined = Nba_combined.reset_index()
Nba_combinedADV = Nba_combinedADV.reset_index()

#Join the 2 files together
Nba_combined= pd.merge(Nba_combined, 
                        Nba_combinedADV[['Player', 'Year', 'WS','VORP','DBPM','PER','BPM','USG%','STL%','TS%','OBPM','DBPM']], 
                        on=['Player', 'Year'], 
                        how='left')

Nba_combined.set_index('Player')
```

```python
#making sure Year is a number
Nba_combined['Year'] = pd.to_numeric(Nba_combined['Year'])


# Helper for True Shooting Attempts
tsa = Nba_combined['FGA'] + 0.44 * Nba_combined['FTA']

# --- Offensive Metrics ---
Nba_combined['PTS_per_TSA'] = Nba_combined['PTS'] / (tsa + 1e-5)
Nba_combined['FT_rate'] = Nba_combined['FTA'] / (Nba_combined['FGA'] + 1e-5)
Nba_combined['3P_rate'] = Nba_combined['3PA'] / (Nba_combined['FGA'] + 1e-5)

# --- Playmaking & Turnovers ---
Nba_combined['AST_TO'] = Nba_combined['AST'] / (Nba_combined['TOV'] + 1e-5)
Nba_combined['TOV%'] = 100 * Nba_combined['TOV'] / (tsa + Nba_combined['TOV'] + 1e-5)

# --- Rebounding & Defense ---
Nba_combined['TRB%_approx'] = 100 * Nba_combined['TRB'] / (Nba_combined['MP'] + 1e-5)
Nba_combined['STL_per_min'] = Nba_combined['STL'] / (Nba_combined['MP'] + 1e-5)
Nba_combined['BLK%'] = 100 * Nba_combined['BLK'] / (Nba_combined['MP'] + 1e-5)

# --- Efficiency ---
Nba_combined['Raw_EFF'] = (
    Nba_combined['PTS'] + Nba_combined['TRB'] + Nba_combined['AST'] +
    Nba_combined['STL'] + Nba_combined['BLK'] -
    (Nba_combined['FGA'] - Nba_combined['FG']) -
    (Nba_combined['FTA'] - Nba_combined['FT']) -
    Nba_combined['TOV']
) / (Nba_combined['MP'] + 1e-5)

# --- Fouls & Discipline ---
Nba_combined['PF_per_min'] = Nba_combined['PF'] / (Nba_combined['MP'] + 1e-5)

# --- Bonus Proxies ---
Nba_combined['Load'] = (
    Nba_combined['FGA'] + 0.44 * Nba_combined['FTA'] +
    Nba_combined['AST'] + Nba_combined['TOV']
)
Nba_combined['Creation_proxy'] = (
    Nba_combined['AST'] * 0.2 + Nba_combined['PTS'] * 0.05) 
```

```python
#set processed data as a dataframe
Nba_combined.to_csv('NbaCombined.csv')
```
